```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from nltk.stem.snowball import SnowballStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
fake = pd.read_csv("fake_news.csv")
true = pd.read_csv("true_news.csv")
```

```python
fake['target'] = 0
true["target"] = 1
```

print(fake.head(5))<br>
print(true.head(5))

```python
data = pd.concat([fake,true],axis=0)
data = data.reset_index(drop=True)
data = data.drop(["subject","date","title"],axis=1)
```

```python
data["text"] = data["text"].apply(word_tokenize)
# print(data["text"].head(5))
```

```python
porter = SnowballStemmer("english")
```

```python
def stem_it(text):
    return [porter.stem(word) for word in text]
```

```python
data["text"] = data["text"].apply(stem_it)
# print(data["text"].head(5))
```

```python
def stop_it(text):
    data = [word for word in text if len(text) > 2]
    return data
data["text"] = data["text"].apply(stop_it)
#print(data["text"].head(5))
```

```python
data["text"] = data["text"].apply(' '.join)
# print(data.head(5))
```

```python
X_train,X_test,y_train,y_test = train_test_split(data["text"], data["target"])
```

```python
tfid = TfidfVectorizer(max_df=0.7)
```

```python
tfidf_train = tfid.fit_transform(X_train)
tfidf_test = tfid.transform(X_test)
```

```python
model_1 = LogisticRegression(max_iter=900)
model_1.fit(tfidf_train, y_train)
pred_1 = model_1.predict(tfidf_test)
accuracy1 = accuracy_score(y_test, pred_1)
print("accuracy of logisticregression is :",accuracy1*100)
```

```python
model_2 = PassiveAggressiveClassifier(max_iter=50)
model_2.fit(tfidf_train,y_train)
pred_2 = model_2.predict(tfidf_test)
accuracy2 = accuracy_score(y_test, pred_2)
print("accuracy of PassiveAggressiveClassifier is :",accuracy2*100)
```
